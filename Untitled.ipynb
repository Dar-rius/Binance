```python
##preprocessing
from  sklearn.impute import MissingIndicator
verifi = MissingIndicator(binance)
print(verifi)
```

```python
#encoder
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
x= binance.drop(['Price(in dollars)', 'Date'], axis=1)
print(x)
y= binance.drop(['Date', 'Open', 'High', 'Low', 'Vol.', 'Change%'], axis=1)
print(y)
X = OrdinalEncoder().fit_transform(x)
print(X.shape)
print(X)
y = LabelEncoder().fit_transform(y)
print(f'La variable de y: {y}')
```

```python
from sklearn.feature_selection import chi2, SelectKBest
chi2(X,y)
select = SelectKBest(chi2, k=4)
select.fit_transform(X,y)
print(select.get_support())
```

```python
X = binance.drop(['Date','Price(in dollars)', 'Vol.' ], axis=1)
X = OrdinalEncoder().fit_transform(X)
print(X)
```

```python
#train and test split
from sklearn.model_selection import train_test_split
#create the variable
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
#pipeline, standardScaler and models
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso, SGDRegressor
#create pipeline
model_A =  make_pipeline(StandardScaler(), SGDRegressor())
model_B =  make_pipeline(StandardScaler(), Lasso())
#fit the models
for model in (model_A, model_B):
    model.fit(X_train, y_train)
    print('les differents models',  model.score(X_test, y_test))
```

```python
def price(ouvert , high, low, change):
    df = pd.DataFrame({
        'Open':[ouvert],
        'High': [high],
        'Low': [low],
        'Change%': [change]
    })
    #encoder
    X = OrdinalEncoder().fit_transform(df)
    #normalisation
    X = StandardScaler().fit_transform(X.squeeze)
    N =  StandardScaler().fit(X.reshape.squeeze)
    prediction = model_A.predict(X)
    print(N.inverse_transform(prediction))
```

```python
print('Bienvenu dans le model de prediction de binance')
ouvert = float(input("Donner le prix d'ouverture: "))
high = float(input('Donner le prix haut: '))
low = float(input('Donner le prix faible: '))
change = float(input('Donner le taux de change: '))

price(ouvert, high, low, change)
```
